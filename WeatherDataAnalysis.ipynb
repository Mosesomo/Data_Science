```python
import pandas as pd

data = pd.read_csv('weather_data.csv')

# print(data.head(20))
weather_clear = data[data['Weather'] == 'Clear']
print(weather_clear.head(10))

no_of_times_wind_speed = data[data['Wind Speed_km/h'] == 4].shape[0]
print(f'\nNumber of times wind speed was 4: {no_of_times_wind_speed}')

checking_null = data.isnull().sum()
print(f'\nChecking null values: \n{checking_null}')

data.rename(columns={'Weather': 'Weather_Condtion'}, inplace=True)
print(f'\nData after renaming: \n{data.head(10)}')

mean_visibility = data['Visibility_km'].mean()
print(f'\nMean visibility: {mean_visibility}')

wind_speed_visibility = data[(data['Wind Speed_km/h'] > 24) & (data['Visibility_km'] == 25)].shape[0]
print(f'\nNumber of times wind speed is greater than 24 and visibility is 25: {wind_speed_visibility}')

mean_by_weather_condition = data.groupby('Weather_Condtion').mean(numeric_only=True)
print(f'\nMean by weather condition: \n{mean_by_weather_condition.head(10)}')

weather_instances = data[(data['Weather_Condtion'] == 'Clear') & (data['Rel Hum_%'] > 50) | (data['Visibility_km'] > 40)]
print(f'\nWeather instances: \n{weather_instances.head(10)}')

number_weather_snow = data[data['Weather_Condtion'] == 'Snow'].shape[0]
print(f'\nNumber of times weather condition is snow: {number_weather_snow}')
```
