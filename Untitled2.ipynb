```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

df = pd.read_csv('weather_data.csv')

print('-------------Data overview and cleaning...\n')
df.info()
duplicated_data = df.duplicated().sum()
print(f'\nduplicated data: {duplicated_data}')

print('\n---------------------Describing the data')
print(df.describe().head())

print('\n---------------------Visualizing the data')
df.hist(figsize=(15, 10), bins=20)
plt.suptitle('Histogram of weather data')
plt.show()

# Distribution plots for key weather parameters
plt.figure(figsize=(15, 10))
plt.subplot(1,3,1)
sns.histplot(df['Temp_C'], kde=True)
plt.title('Temperature Distribution')

plt.subplot(1,3,2)
sns.histplot(df['Rel Hum_%'], kde=True)
plt.title('Humidity Distribution')

plt.subplot(1,3,3)
sns.histplot(df['Wind Speed_km/h'], kde=True)
plt.title('Wind Speed Distribution')

plt.tight_layout()
plt.show()

df['Date/Time'] = pd.to_datetime(df['Date/Time'])
df.set_index('Date/Time', inplace=True)

plt.figure(figsize=(15, 10))
plt.subplot(3, 1, 1)
df['Temp_C'].plot()
plt.title('Temperature Over Time')

plt.subplot(3, 1, 2)
df['Rel Hum_%'].plot()
plt.title('Humidity Over Time')

plt.subplot(3, 1, 3)
df['Wind Speed_km/h'].plot()
plt.title('Wind Speed Over Time')

plt.tight_layout()
plt.show()

numerical_df = df.select_dtypes(include=['float', 'int'])
corr_matrix = numerical_df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()

# Extracting month and year
df['Month'] = df.index.month
df['Year'] = df.index.year

# Group by month and calculate mean temperature and humidity
monthly_data = df.groupby('Month')[['Temp_C', 'Rel Hum_%']].mean()

# Plotting monthly data
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
sns.lineplot(data=monthly_data, x=monthly_data.index, y='Temp_C')
plt.title('Average Temperature by Month')

plt.subplot(1, 2, 2)
sns.lineplot(data=monthly_data, x=monthly_data.index, y='Rel Hum_%')
plt.title('Average Humidity by Month')

plt.tight_layout()
plt.show()

# Anomalies in temperature and humidity
temperature_anomalies = df[(df['Temp_C'] > df['Temp_C'].quantile(0.99)) | (df['Temp_C'] < df['Temp_C'].quantile(0.01))]
print("Temperature Anomalies:\n", temperature_anomalies.head(10))

humidity_anomalies = df[(df['Rel Hum_%'] > df['Rel Hum_%'].quantile(0.99)) | (df['Rel Hum_%'] < df['Rel Hum_%'].quantile(0.01))]
print("Humidity Anomalies:\n", humidity_anomalies.head(10))

```
